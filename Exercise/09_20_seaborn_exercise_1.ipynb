````python
raw_data = {'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
            'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
            'female': [0, 1, 1, 0, 1],
            'age': [42, 52, 36, 24, 73],
            'preTestScore': [4, 24, 31, 2, 3],
            'postTestScore': [25, 94, 57, 62, 70]}

df=pd.DataFrame(raw_data)
df.head()
````

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

````python
plt.figure(figsize=(10,6))
sns.scatterplot(x='preTestScore', y='postTestScore', data=df, size=df['age'])
plt.xlabel('preTestScore')
plt.ylabel('postTestScore')
plt.show()
````

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

````python
plt.figure(figsize=(10,6))
sns.scatterplot(x='preTestScore', y='postTestScore', data=df, size=df['postTestScore']*4.5, hue=df['female'])
plt.xlabel('preTestScore')
plt.ylabel('postTestScore')

plt.show()
````

### BONUS: Create your own question and answer it.

preTestScore과 postTestScore의 violin plot을 각각 하나의 figure안에 생성하여 분포를 비교하라.

````python
fig, ax = plt.subplots(nrows = 1, figsize = (6, 5), ncols = 2)

sns.violinplot(data = df, y = "preTestScore", ax=ax[0])
plt.ylabel('preTestScore')
sns.violinplot(data = df, y = "postTestScore", ax=ax[1])
plt.ylabel('postTestScore')

plt.show()
````
